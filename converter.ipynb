```python
import os
from tqdm.notebook import tqdm
from pydub import AudioSegment
from joblib import Parallel, delayed
```

```python
# File paths
main_folder_path = "/content/drive/MyDrive/Colab Notebooks/aud-ml-datasets/"


# WAV file paths
wav_folder_name ="wav-audios"
wav_folder_path = f"{main_folder_path}{wav_folder_name}"

# creating the WAV folder if it doesnot exist
# if not os.path.exists(wav_folder_path):
#   os.mkdir(wav_folder_path)


# MP3 file path
mp3_folder_name = "Folder_2"
mp3_folder_path = f"{main_folder_path}{mp3_folder_name}"
mp3_audio_files = os.listdir(mp3_folder_path)
```

```python
# Converting MP3 files to WAV form

def convert_mp3_wav(filename):

  path = f"{mp3_folder_path}/{filename}"
  save_path = f"{wav_folder_path}"
  wav_output = f"{save_path}/{filename[:-4]}.wav"

  if not os.path.exists(save_path):
    os.makedirs(save_path, exist_ok=True)

  if os.path.exists(path):
    try:
      sound = AudioSegment.from_mp3(path)
      sound = sound.set_frame_rate(16000)
      sound.export(f"{wav_output}", format="wav")

    except:
      print(f"an error occured for file {filename}")


# print(os.listdir(f"{main_folder_path}"))

print(mp3_audio_files)
```

```python
# Running paralell jobs to convert mp3 files faster

audio_mp3_files = os.listdir(mp3_folder_path)
# print(audio_mp3_files)

Parallel(n_jobs=8, backend="multiprocessing")(delayed(convert_mp3_wav)(filename) for filename in tqdm(audio_mp3_files))
```
